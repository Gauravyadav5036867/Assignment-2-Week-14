```python
#Question 1

import pandas as pd

# Load the flight price dataset
flight_data = pd.read_csv('flight_price_dataset.csv')

# Display the dimensions of the dataset
print("Number of rows:", flight_data.shape[0])
print("Number of columns:", flight_data.shape[1])
```

```python
#Question 2

import matplotlib.pyplot as plt

# Assuming 'flight_data' contains the loaded dataset with a column named 'Price'
flight_prices = flight_data['Price']

# Create a histogram to visualize the distribution of flight prices
plt.figure(figsize=(10, 6))
plt.hist(flight_prices, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Question 3

# Calculate the minimum and maximum prices
min_price = flight_prices.min()
max_price = flight_prices.max()

# Calculate the range of prices
price_range = max_price - min_price

print("Minimum price:", min_price)
print("Maximum price:", max_price)
print("Price range:", price_range)
```

```python
#Question 4

import seaborn as sns

# Assuming 'flight_data' contains the loaded dataset with columns 'Airline' and 'Price'
plt.figure(figsize=(12, 8))
sns.boxplot(x='Airline', y='Price', data=flight_data, palette='viridis')
plt.title('Price of Flights by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Question 5 

# Create a boxplot to identify potential outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x='Price', data=flight_data, palette='pastel')
plt.title('Boxplot of Flight Prices')
plt.xlabel('Price')
plt.show()
```

```python
#Question 9 

import pandas as pd

# Load the Google Playstore dataset
playstore_data = pd.read_csv('google_playstore.csv')

# Display the dimensions of the dataset
print("Number of rows:", playstore_data.shape[0])
print("Number of columns:", playstore_data.shape[1])
```

```python
#uestion 10 

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'playstore_data' contains the loaded dataset with columns 'Category' and 'Rating'
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Rating', data=playstore_data, palette='Set3')
plt.title('Rating of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
#Question 11

# Check for missing values in the dataset
missing_values = playstore_data.isnull().sum()

print("Missing values in the dataset:")
print(missing_values)
```

```python
#Question 12

import matplotlib.pyplot as plt

# Assuming 'playstore_data' contains the loaded dataset with columns 'Size' and 'Rating'

# Convert 'Size' column to numeric by removing 'M' and 'k' and converting to MB
playstore_data['Size'] = playstore_data['Size'].str.replace('M', '').str.replace('k', '').astype(float)
playstore_data.loc[playstore_data['Size'] < 1, 'Size'] *= 1000  # Convert smaller sizes to MB

# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(playstore_data['Size'], playstore_data['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size (MB)')
plt.ylabel('Rating')
plt.grid(True)
```

```python
#Question 13

import matplotlib.pyplot as plt

# Assuming 'playstore_data' contains the loaded dataset with columns 'Type' and 'Price'

# Filter out free apps (Price == 0)
paid_apps = playstore_data[playstore_data['Price'] != '0']

# Convert 'Price' column to numeric by removing '$' and converting to float
paid_apps['Price'] = paid_apps['Price'].str.replace('$', '').astype(float)

# Calculate average price by app type
average_price_by_type = paid_apps.groupby('Type')['Price'].mean()

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(8, 6))
average_price_by_type.plot(kind='bar', color='skyblue')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
#Question 14

# Assuming 'playstore_data' contains the loaded dataset with columns 'App' and 'Installs'

# Remove any commas and plus signs from the 'Installs' column and convert to numeric
playstore_data['Installs'] = playstore_data['Installs'].str.replace(',', '').str.replace('+', '').astype(int)

# Create a frequency table of app installations
install_frequency_table = playstore_data.groupby('App')['Installs'].sum().sort_values(ascending=False)

# Display the top 10 most popular apps
top_10_apps = install_frequency_table.head(10)
print("Top 10 Most Popular Apps:")
print(top_10_apps)
```
